### Dataset 1: Dog Population of the City of Zurich (kul100od1001.csv)
The data contains information on both dog owners and their dogs, including demographic details such as age, gender, and geographic location of the owner, as well as characteristics of the dogs themselves, including breed, age, gender, and color. Additionally, it includes indicators for whether the dog is a mixed breed, as well as the number of dogs owned by each individual.

#### Reference
Hundekontrolle, Stadtpolizei Zürich, Sicherheitsdepartement. (2015, September 24). *Hundebestände der Stadt Zürich seit 2015* (Version fe117e1a-573a-47fe-b304-65cedbbb0e95@stadt-zurich). Open Data Zürich. [https://data.stadt-zuerich.ch/dataset/sid_stapo_hundebestand_od1001](https://data.stadt-zuerich.ch/dataset/sid_stapo_hundebestand_od1001)

#### Data Size and Attribute Types
The dataset consists of 80,788 entries, with 32 attributes that are categorical or ordinal, including some redundancies due to the pre-coding of the categories.

#### Variable Descriptions
- **StichtagDatJahr**: The year of the relevant reference date (likely the year the data was recorded).
- **DatenstandCd**: A code for the data status (probably indicating the version or timestamp of the data collection).
- **HalterId**: A unique ID for the dog owner.
- **AlterV10Cd**, **AlterV10Lang**, **AlterV10Sort**: Information on the age of the dog owner. "Cd" may represent a code, "Lang" the full description, and "Sort" a sorting category.
- **SexCd**, **SexLang**, **SexSort**: Gender information for the dog owner (code, text, sorting order).
- **KreisCd**, **KreisLang**, **KreisSort**: Information on the geographic district where the owner resides (code, name, sorting order).
- **QuarCd**, **QuarLang**, **QuarSort**: Information on the owner’s neighborhood (code, name, sorting order).
- **Rasse1Text**, **Rasse2Text**: The breed of the dog, where "Rasse2Text" indicates a second breed for mixed or double breeds.
- **RasseMischlingCd**, **RasseMischlingLang**, **RasseMischlingSort**: Indicator of whether the dog is a mixed breed (code, text, sorting).
- **RassentypCd**, **RassentypLang**, **RassentypSort**: A classification of the dog by breed type (e.g., working dog, family dog), potentially based on a standard categorization.
- **GebDatHundJahr**: The dog's birth year.
- **AlterVHundCd**, **AlterVHundLang**, **AlterVHundSort**: Information on the dog’s age (code, text, and sorting).
- **SexHundCd**, **SexHundLang**, **SexHundSort**: Gender information for the dog (code, text, sorting order).
- **HundefarbeText**: The color of the dog.
- **AnzHunde**: The number of dogs the owner possesses.

#### Dataset 2: Equivalized Household Income of Taxable Natural Persons by Household Type and City District (wir100od100b.csv)
The data contains information on equivalized household income for taxable natural persons in Zurich, categorized by household type and city district. It includes median, 25th percentile, and 75th percentile income values, as well as codes and descriptions for the district and household type, collected for the years 2013 to 2021.

#### Reference
Steueramt, Finanzdepartement. (2022, November 1). *Haushaltsäquivalenz-Einkommen steuerpflichtiger natürlicher Personen nach Haushaltstyp und Stadtkreis* (Version 1.0). Statistik Stadt Zürich, Präsidialdepartement. Open Data Zürich. [https://data.stadt-zuerich.ch/dataset/fd_median_hhaequeinkommen_kreis_od100b](https://data.stadt-zuerich.ch/dataset/fd_median_hhaequeinkommen_kreis_od100b)

#### Data Size and Attribute Types
The dataset consists of 1,272 entries and 10 attributes. It contains categorical attributes for the district and numeric data on equivalized household income.

#### Variable Descriptions
- **StichtagDatJahr**: The year for which the tax was assessed.
- **KreisSort**: The sorting code for the district of the current residence.
- **KreisCd**: The district code of the current residence.
- **KreisLang**: The name of the district of the current residence.
- **HHtypSort**: The sorting code for the household type.
- **HHtypCd**: The sorting code for the household type.
- **HHtypLang**: The description of the household type.
- **aequEK_p50**: The median value of equivalized household income for taxable natural persons in Zurich, in thousands of Swiss Francs.
- **aequEK_p25**: The 25th percentile value of equivalized household income for taxable natural persons in Zurich, in thousands of Swiss Francs.
- **aequEK_p75**: The 75th percentile value of equivalized household income for taxable natural persons in Zurich, in thousands of Swiss Francs.

```python
def plot_top_bottom_categories(df, column, top_n=10, rarest=False):
    """
    Create a bar plot for the top or bottom N categories in a chosen variable.
    
    Parameters:
    df (DataFrame): The dataframe containing the data.
    column (str): The column name to plot.
    top_n (int): The number of top or bottom categories to plot.
    rarest (bool): If True, plot the rarest categories. If False, plot the most frequent categories.
    """
    if rarest:
        categories = df[column].value_counts().nsmallest(top_n)
        title = f'Top {top_n} Rarest Categories in {column}'
    else:
        categories = df[column].value_counts().nlargest(top_n)
        title = f'Top {top_n} Most Frequent Categories in {column}'
    
    df_filtered = df[df[column].isin(categories.index)]
    
    plt.figure(figsize=(8, 3))
    sns.countplot(data=df_filtered, x=column, color="skyblue", order=categories.index)
    plt.title(title, fontsize=14, weight='bold')
    plt.xlabel(variable_descriptions[column], fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=90, fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, axis='y', linestyle='--', linewidth=0.7)
    sns.despine(left=True, bottom=True)
    plt.show()

# Dog Colors
plot_top_bottom_categories(df_dogs_filtered, column='HundefarbeText', top_n=10, rarest=False)
plot_top_bottom_categories(df_dogs_filtered, column='HundefarbeText', top_n=10, rarest=True)

# God breeds
plot_top_bottom_categories(df_dogs_filtered, column='Rasse1Text', top_n=10, rarest=False)
plot_top_bottom_categories(df_dogs_filtered, column='Rasse1Text', top_n=10, rarest=True) 
```
